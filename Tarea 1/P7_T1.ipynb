```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Euler(k,u0,step):
    t = np.arange(0,k)
    u = np.zeros(k)
    ui = u0
    for i in range(0,k):
        u[i] = ui*(1-step)**i
        ui = u[i]
    return t,u
```

```python
a = Euler(100,1,1.1)
b = Euler(100,1,1.5)
c = Euler(100,1,1.9)

expy = np.exp(-a[0])
```

```python
plt.plot(a[0],a[1],label="Step de 1.1")
plt.plot(b[0],b[1],label="Step de 1.5")
plt.plot(c[0],c[1],label="Step de 1.9")
plt.plot(a[0],expy,label="Funci√≥n exponencial")

leg = plt.legend(loc='upper center')

plt.xlabel("Tiempo (s)")
plt.ylabel("U")

plt.show()
```
