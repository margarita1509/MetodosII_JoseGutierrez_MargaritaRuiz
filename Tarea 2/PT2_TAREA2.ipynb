```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dif(x,y):
    return x*y**2 - 2*y/x - 1/(x**3)
```

```python
def Solucion(x):
    return 1/(x**2) - 1/(1+((x**2)/2))
```

```python

x = np.linspace(np.sqrt(2),100,250)
y = Solucion(x)

h = x[1] - x[0]
```

```python
def GetEuler(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = dif(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1
```

```python
y1 = np.zeros_like(x)
y1[0] = 0
GetEuler(x,y1,h)
```

```python
plt.plot(x,y,label="Solucion analítica")
plt.plot(x,y1,label="Solución numérica")
plt.legend()

plt.xlabel("x")
plt.ylabel("y")

```
