```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dif(x):
    return (-x)*(np.pi**2)
```

```python
def f(t):
    a = 1/np.pi
    return a*np.sin(np.pi*t)
```

```python
def Verlet(t,x):
    h = t[1] - t[0]
    v = 1 + dif(x[0])*h    
    x[1] = x[0] + v*h
    
    for i in range(2,len(t)):
        r = dif(x[i-1])
        x[i] = 2*x[i-1] - x[i-2] + r*(h**2)
        
```

```python
t = np.linspace(0,5,50)
x = np.zeros(len(t))

Verlet(t,x)
x_new = f(t)
```

```python
plt.plot(t,x,label="Solución con Verlet")
plt.plot(t,x_new,label="Solución teórica")

plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia (m)")

plt.legend()
```
