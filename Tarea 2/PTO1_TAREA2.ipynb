```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#Elegimos una ecuaci贸n diferencial para estimar la soluci贸n
def f(x,y):
    return x + 2*y
```

```python
def Solucion(x):
    return -0.5*x - (1-np.exp(2*x))/4.
```

```python
#ELegimos un paso para la funci贸n
N = 5
t = np.linspace(0,1,N)
```

```python
h = t[1] - t[0]
```

```python
#creamos la soluci贸n real
yext = Solucion(t)
```

```python
plt.scatter(t,yext)
```

```python
#Implementamos Runge-Kutta 1
def GetEuler(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1


y1 = np.zeros_like(t)
y1[0] = 0

GetEuler(t,y1,h)

plt.plot(t,yext)
plt.plot(t,y1)
```

```python
#Implementamos Runge-Kutta 2
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)

def f(x,y):
    return x + 2*y

y2 = np.zeros_like(t)
y2[0] = 0.

GetEuler2(t,y2,h)

plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y2,label='Euler2')
plt.legend()
```

```python
#Implementamos Runge-Kutta 3
def GetEuler3(x,y,h):
    for i in range(1,len(x)):
        k1=f(x[i-1],y[i-1])
        k2=f(x[i-1]+0.5*h,y[i-1]+0.5*h*k1)
        k3=f(x[i-1]+h,y[i-1]-h*k1+2*h*k2)
        y[i]=y[i-1]+(k1+4*k2+k3)*h/6
def f(x,y):
    return x+2*y

y3=np.zeros_like(t)
y3[0]=0.

GetEuler3(t,y3,h)

plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y2,label='Euler2')
plt.plot(t,y3,label='Euler3')
plt.legend()
```

```python
#Implementamos Runge-Kutta 4
def GetEuler4(x,y,h):
    for i in range(1,len(x)):
        k1=f(x[i-1],y[i-1])
        k2=f(x[i-1]+0.5*h,y[i-1]+0.5*k1*h)
        k3=f(x[i-1]+0.5*h,y[i-1]+0.5*h*k2)
        k4=f(x[i-1]+h,y[i-1]+k3*h)
        y[i]=y[i-1]+(k1+2*k2+2*k3+k4)*h/6
    
def f(x,y):
    return x+2*y

y4=np.zeros_like(t)
y4[0]=0.

GetEuler4(t,y4,h)

plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y2,label='Euler2')
plt.plot(t,y3,label='Euler3')
plt.plot(t,y4,label='Euler4')
plt.legend()
print('N贸tese que para la ecuaci贸n diferencial establecida, al reducir el paso se nota que para una menor cantidad de iteraciones, la estimaci贸n de Runge-Kutta mejora a medida que mejora el grado de Runge-Kutta que estemos implementando, variando el paso vimos que, en N=10 desde Runge-Kutta 2 es indistinguible, al menos para esta ecuaci贸n, de entre el 3 y el 4. Esto implica que al desarrollar m茅todos de Runge-Kutta de mayor orden, se puede optimizar el proceso de estimaci贸n para iteraciones peque帽as al rededor de la funci贸n que se busca estabilizar, pues el m茅todo tiende a oscilar al rededor de esta.')
```
