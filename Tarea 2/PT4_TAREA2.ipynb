```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dif(x,y,q):
    return y**q
```

```python
def Solucion(x,q):
    y = np.e**x
    if q < 1:
        y = (x*(1-q)+1)**(1/(1-q))
    return y
```

```python
x = np.linspace(0,100,250)
y1 = Solucion(x,0)
y2 = Solucion(x,0.2)
y3 = Solucion(x,0.4)
y4 = Solucion(x,0.7)
y5 = Solucion(x,0.9)
y6 = Solucion(x,1)


h = x[1] - x[0]
```

```python
def GetEuler(x,y,h,q):
    
    for i in range(1,len(x)):
        
        k1 = dif(x[i-1],y[i-1],q)
        y[i] = y[i-1] + h*k1
```

```python
ya = np.zeros_like(x)
ya[0] = 1

yb = np.zeros_like(x)
yb[0] = 1

yc = np.zeros_like(x)
yc[0] = 1

yd = np.zeros_like(x)
yd[0] = 1

ye = np.zeros_like(x)
ye[0] = 1

yf = np.zeros_like(x)
yf[0] = 1

GetEuler(x,ya,h,0)
GetEuler(x,yb,h,0.2)
GetEuler(x,yc,h,0.4)
GetEuler(x,yd,h,0.7)
GetEuler(x,ye,h,0.9)
GetEuler(x,yf,h,1)
```

```python
plt.plot(x,y1,label="Solución analítica")
plt.legend()
plt.plot(x,ya, label="Solución numérica")
plt.legend()

plt.xlabel("U")
plt.ylabel("t")
```
